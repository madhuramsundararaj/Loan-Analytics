```python
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv('cs-training.csv')
print("Loaded:", df.shape)

# Fix missing values
df['MonthlyIncome'] = df['MonthlyIncome'].fillna(df['MonthlyIncome'].median())
df['NumberOfDependents'] = df['NumberOfDependents'].fillna(df['NumberOfDependents'].mode()[0])

# Verify all zeros
print(df.isnull().sum())
```

```python
# Remove impossible values
df = df[df['age'] > 18]
df = df[df['age'] < 100]
df = df[df['MonthlyIncome'] >0]
df = df[df['RevolvingUtilizationOfUnsecuredLines'] < 1]

# Check how many rows remain
print("Rows after cleaning:", df.shape[0])
```

```python
# Feature 1: EMI burden ratio
# How much debt compared to income
df['emi_burden'] = df['DebtRatio'] / (df['MonthlyIncome'] + 1)

# Feature 2: Risk Score
# Combined score using two key risk columns
df['risk_score'] = (df['DebtRatio'] * 0.4 + df['RevolvingUtilizationOfUnsecuredLnes'] * 0.6)

# Feature 3: Age group
df['age_group'] = pd.cut(df['age'], 
                         bins=[0, 30, 50, 100], 
                         Labels=['Young', 'Mid', 'Senior'])

# Confirm new columns exist
print(df[['emi_burden', 'risk_score', 'age_group']].head())
```
